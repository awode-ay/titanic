```python
open_f = open("titanic_train.csv")
import csv
read_f = csv.reader(open_f)
lines = list(read_f)
header = lines[0]
print (header)
p_info = lines[1:]
survived = []
not_survived = []
pclass = []
male_survivors = []
female_survivors = []
age = []
for info in p_info:
    if info[1] == "1":
        survived.append(info[1])
        if info[4] == "male":
            male_survivors.append(info[4])
        else:
            female_survivors.append(info[4])
    else:
        not_survived.append(info[1])
    ages = info[5]
    age.append(ages) 
gender_survivors = ["Male","Female"]
no_of_surv = [len(male_survivors), len(female_survivors)] 
import matplotlib.pyplot as plt
plt.xlabel("The gender of survivors")
plt.ylabel("The number of survivors")
plt.title("Grouping of titanic survivors by gender")
plt.bar(gender_survivors,no_of_surv, color = "green" )

plt.show()
```

```python
plt.title("Grouping of titanic survivors by gender")
plt.pie(no_of_surv, labels = gender_survivors, autopct="%1.1f%%", colors = ["lightgreen","lightblue"])
plt.show()
```

```python
survival = ["survived","not survived"]
no_of_sur = [len(survived), len(not_survived)]
plt.title("Survival Rate of Titanic Passengers")
plt.pie(no_of_sur, labels = survival, autopct="%1.1f%%", colors = ["pink","orange"])
plt.show()
```
